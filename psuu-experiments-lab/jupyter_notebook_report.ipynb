- **Design**: 
The purpose of the automated PSUU process is to search the parameter space for a cadCAD model in a structured way, focusing primarily on areas that fulfill requirements for high performance. We envision the following steps:
0.a. Fix a set of states that will be used in the cadCAD model. 
0.b. Generate an initial parameter grid to begin, according to some **defined structure.**
1. Using the current paramter grid, run an experiment, recording KPIs.
2. Process the KPIs into a single **ranking measurement** to be used in determining the importance of parameters.
3. Using the top $m$ KPIs according to the ranking in Step 2, update the parameter grid according to an **updating rule**. 

We will have more to say about the bolded words in the section below. 

- **Defined Structure:** 
For the purposes of an initial prototype, we create a `TypedDict` object which associates to each parameter value to be tested a tuple consisting of `(lower_bound, upper_bound)`. Having just two parameter values is more than enough to generate a large number of experiments, and offers a straightforward approach for updating after features are ranked. 

- **Ranking Methods:**
With a large number of parameters, the computation for updating every pair of parameter values quickly becomes computationally infeasible. Accordingly, we choose only specific parameters to update in the parameter grid at each step. 

In the specific situation that originates this study, we have a processing step where KPIs are processed using *threshold inequalities* to a 0-1 vector, indicating whether or not specific requirements were met. The goal is to rank parameters such that we can find the region of the parameter space where as many threshold inequalities as possible are satisfied. 

We would like to test two possibilities:
1. Feature Importance Using Random Forest (based on the `cadCAD-machine-search` repository). 8 view the correct approach here as using [Multi-Output Classification](https://scikit-learn.org/stable/modules/multiclass.html), and many approaches besides Random Forest are possible. 8 have concerns here, which I will raise in the appropriate venue. 
2. `mutual_info_classification`: For each parameter, calculate the mutual information between the feature and the binary variable of the threshold inequality. 

There are many feature importance methods that could be used, and the two above are just a starting point. 

- **Updating Rule**

For a given `(lower_bound, upper_bound)` pair, we will replace either `lower_bound` or `upper_bound` with a new value. We implement a method proposed by Dr. Jamsheed Shoresh, reminiscent of Bisection Method for bounding the solutions of root-finding problems.
- we calculate `new_bound` as the average of `upper_bound` and `lower_bound`
- if `lower_bound` has better performance, the new parameter pair is `(lower_bound, new_bound)`
- if `upper_bound` has better performance, the new parameter pair is `(new_bound, upper_bound)`

Other update rules are possible, but this one is a straightforward starting point. 



- **Key Questions**:
There are two key questions related to the PSUU, each of which needs to be tested.
**Question 1.** Is the method practical, in terms of performance time? The amount of time taken will be the same as an open-loop pre-selected experiment process in terms of cadCAD simulation time and data processing. The parameter grid processing steps add computational overhead, and the question is whether this additional time makes the process infeasible. 

**Question 2.** Which **ranking method** should be used to identify the parameters to be updated at each time step?
